```python
# importing Important Library for Continuous distribution and DIscrete Distribution Distribution 
from scipy import stats as st
#Library to impliment Uniform Distribution
from scipy.stats import uniform
#Library to Impliment Binomial Distribution
from scipy.stats import binom
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

# **Generating random samples Iteratively for Continuous Population Distribution And FInding Several Statistics for samples along with plotting the resulting sampling distribution, mean and variance of the statistics** 

```python
#Generating random numbers from uniform distribution
for k in range(10):
  # Creating a Variable for sample
  continuous_sample_count = 10000
  #Variable that contain start point for Random Number generation
  start = 10
  width = 20
  #Creating an Object that Store the generated sample 
  data_uniform = uniform.rvs(size=continuous_sample_count, loc = start, scale=width)
  plt.figure(k)
  #Plotting the Graph
  ax = sns.distplot(data_uniform,bins=100,kde=True,color='skyblue',hist_kws={"linewidth": 15,'alpha':1})
  ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
  print("Iteration", k)
  #Finding the mean
  display(np.mean(data_uniform))
  #Finding The median
  display(np.median(data_uniform))
  #Finding the Variance
  display(np.var(data_uniform))
  #Finding the Standard Deviation
  display(np.std(data_uniform))
  print("")
```

# **Generating random samples Iteratively for Discrete Population Distribution And FInding Several Statistics for samples along with plotting the resulting sampling distribution, mean and variance of the statistics** 

```python
# random numbers from discrete distribution
for k in range(10):
  #Creating a Variable for Discrete Sample Count
  discrete_sample_count = 10000
  #Creating a Variable for No of Trial
  n=10
  #Creating a variablr for Binomial Parameter that tells probability of Success for one trial
  p=0.5
  #Creating a nobject that store the generated sample
  data_discrete = binom.rvs(n, p, size=discrete_sample_count)
  plt.figure(k)
  #Plotting the graph
  ax = sns.distplot(data_discrete,bins=10,kde=True,color='red',hist_kws={"linewidth": 15,'alpha':1})
  ax.set(xlabel='Discrete Distribution ', ylabel='Frequency')
  print("iteration",k)
  #Finding the Mean
  display(np.mean(data_discrete))
  #Finding the median
  display(np.median(data_discrete))
  #Finding the Standard Deviation
  display(np.std(data_discrete))
  #Finding the Variance
  display(np.var(data_discrete))
  print(" ")
```

**Implimenting One Tailored Test for Continuous Distribution**

```python
st.ttest_1samp(data_uniform, 20)
#P value>0.05 therefore failed to reject
```

**Implimenting One Tailored Test for Discrete Distribution**

```python
st.ttest_1samp(data_discrete, 5)
#P value>0.05 therefore failed to reject
```
